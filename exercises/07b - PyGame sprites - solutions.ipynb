## Esercizio 1 - Crea e visualizza 2 sprite

Prova a replicare la seguente schermata:

```python
import pygame
import random

class Player(pygame.sprite.Sprite):
    def __init__(self, center):
        super(Player, self).__init__()
        self.image = pygame.image.load("/home/matteo/Downloads/Knight.png").convert_alpha()
        self.rect = self.image.get_rect(center=center)

class Enemy(pygame.sprite.Sprite):
    def __init__(self, center):
        super(Enemy, self).__init__()
        self.image = pygame.image.load("/home/matteo/Downloads/Bomb Person.png").convert_alpha()
        self.rect = self.image.get_rect(center=center)

pygame.init()

screen = pygame.display.set_mode((500, 500))

clock = pygame.time.Clock()
FPS = 30

player = Player((
    random.randint(0, 500),
    random.randint(0, 500)
))

enemy = Enemy((
    random.randint(0, 500),
    random.randint(0, 500)
))

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill("white")

    screen.blit(player.image, player.rect)
    screen.blit(enemy.image, enemy.rect)

    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```

## Esercizio 2 - Update sprite
Aggiungi agli sprite precedenti una funzione update che permetta allo sprite dell'utente il movimento da tastiera, mentre lo sprite che rappresenta il nemico dovrà muoversi ad ogni istante verso l'utente.  
Se non riesci a implementare la logica del movimento automatico del nemico, non ti preoccupare. Concentrati sul fare in modo che il personaggio dell'utente si muova in base ai tasti premuti.  
<img src="resources/07b-es2.gif" alt="schermata esercizio 2" width=350>  

```python
import pygame
import random
import math

from pygame.locals import(
    K_DOWN,
    K_UP,
    K_LEFT,
    K_RIGHT
)

class Player(pygame.sprite.Sprite):
    def __init__(self, center, speed = 5):
        super(Player, self).__init__()
        self.image = pygame.image.load("/home/matteo/Downloads/Knight.png").convert_alpha()
        self.rect = self.image.get_rect(center=center)
        self.speed = speed

    def update(self, pressed_keys):
        if pressed_keys[K_DOWN]:
            self.rect.move_ip(0, self.speed)
        if pressed_keys[K_UP]:
            self.rect.move_ip(0, -self.speed)
        if pressed_keys[K_LEFT]:
            self.rect.move_ip(-self.speed, 0)
        if pressed_keys[K_RIGHT]:
            self.rect.move_ip(self.speed, 0)

class Enemy(pygame.sprite.Sprite):
    def __init__(self, center, speed = 4):
        super(Enemy, self).__init__()
        self.image = pygame.image.load("/home/matteo/Downloads/Bomb Person.png").convert_alpha()
        self.rect = self.image.get_rect(center=center)
        self.speed = speed

    def update(self, target_point):
        current_point = self.rect.center
        delta = (
            target_point[0] - current_point[0],
            target_point[1] - current_point[1]
        )
        delta_len = math.sqrt(delta[0]**2 + delta[1]**2)
        if delta_len == 0:
            speed_vector = (0, 0)
        else:
            speed_vector = (
                delta[0] / delta_len * self.speed,
                delta[1] / delta_len * self.speed
            )
        self.rect.move_ip(speed_vector[0], speed_vector[1])

pygame.init()

screen = pygame.display.set_mode((500, 500))

clock = pygame.time.Clock()
FPS = 30

player = Player((
    random.randint(0, 500),
    random.randint(0, 500)
))

enemy = Enemy((
    random.randint(0, 500),
    random.randint(0, 500)
))

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill("white")

    pressed_keys = pygame.key.get_pressed()

    player.update(pressed_keys)
    enemy.update(player.rect.center)

    screen.blit(player.image, player.rect)
    screen.blit(enemy.image, enemy.rect)

    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```

## Esercizio 3 - Rilevamento contatto
Aggiungi una logica che permetta di determinare il contatto tra nemico e utente. Se l'utente viene toccato dal nemico, uccidi l'utente.  
Se non riesci a rimuovere l'utente dalla visualizzazione non ti preoccupare. Mostra un contatore sullo schermo che conti quante volte utente e nemico si sono toccati.

```python
import pygame
import random
import math

from pygame.locals import(
    K_DOWN,
    K_UP,
    K_LEFT,
    K_RIGHT
)

class Player(pygame.sprite.Sprite):
    def __init__(self, center, speed = 5):
        super(Player, self).__init__()
        self.image = pygame.image.load("/home/matteo/Downloads/Knight.png").convert_alpha()
        self.rect = self.image.get_rect(center=center)
        self.speed = speed

    def update(self, pressed_keys):
        if pressed_keys[K_DOWN]:
            self.rect.move_ip(0, self.speed)
        if pressed_keys[K_UP]:
            self.rect.move_ip(0, -self.speed)
        if pressed_keys[K_LEFT]:
            self.rect.move_ip(-self.speed, 0)
        if pressed_keys[K_RIGHT]:
            self.rect.move_ip(self.speed, 0)

class Enemy(pygame.sprite.Sprite):
    def __init__(self, center, speed = 4):
        super(Enemy, self).__init__()
        self.image = pygame.image.load("/home/matteo/Downloads/Bomb Person.png").convert_alpha()
        self.rect = self.image.get_rect(center=center)
        self.speed = speed

    def update(self, target_point):
        current_point = self.rect.center
        delta = (
            target_point[0] - current_point[0],
            target_point[1] - current_point[1]
        )
        delta_len = math.sqrt(delta[0]**2 + delta[1]**2)
        if delta_len == 0:
            speed_vector = (0, 0)
        else:
            speed_vector = (
                delta[0] / delta_len * self.speed,
                delta[1] / delta_len * self.speed
            )
        self.rect.move_ip(speed_vector[0], speed_vector[1])

pygame.init()

screen = pygame.display.set_mode((500, 500))

clock = pygame.time.Clock()
FPS = 30

player = Player((
    random.randint(0, 500),
    random.randint(0, 500)
))

enemy = Enemy((
    random.randint(0, 500),
    random.randint(0, 500)
))

enemies = pygame.sprite.Group()
enemies.add(enemy)

all_sprites = pygame.sprite.Group()
all_sprites.add(player)
all_sprites.add(enemies)

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill("white")

    pressed_keys = pygame.key.get_pressed()

    if (player.alive):
        player.update(pressed_keys)
    enemy.update(player.rect.center)

    collision = pygame.sprite.spritecollide(player, enemies, False)
    if len(collision):
        player.kill()

    all_sprites.draw(screen)

    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```

## Esercizio 4 - Mouse
Aggiungi una logica che permetta di leggere i click del mouse. Se l'utente clicca sul nemico elimina il nemico.  
**SUGGERIMENTO**: puoi rilevare se un click è stato fatto sul nemico controllando se il punto su cui l'utente clicca sta dentro al rettangolo occupato dal nemico.  
Usa questa funzione:
`enemy.rect.collidepoint(point)`  
[Documentazione](https://pyga.me/docs/ref/rect.html#pygame.Rect.collidepoint)  
<img src="resources/07b-es4.gif" alt="schermata esercizio 4" width=350>  

```python
import pygame
import random
import math

from pygame.locals import(
    K_DOWN,
    K_UP,
    K_LEFT,
    K_RIGHT
)

class Player(pygame.sprite.Sprite):
    def __init__(self, center, speed = 5):
        super(Player, self).__init__()
        self.image = pygame.image.load("/home/matteo/Downloads/Knight.png").convert_alpha()
        self.rect = self.image.get_rect(center=center)
        self.speed = speed

    def update(self, pressed_keys):
        if pressed_keys[K_DOWN]:
            self.rect.move_ip(0, self.speed)
        if pressed_keys[K_UP]:
            self.rect.move_ip(0, -self.speed)
        if pressed_keys[K_LEFT]:
            self.rect.move_ip(-self.speed, 0)
        if pressed_keys[K_RIGHT]:
            self.rect.move_ip(self.speed, 0)

class Enemy(pygame.sprite.Sprite):
    def __init__(self, center, speed = 4):
        super(Enemy, self).__init__()
        self.image = pygame.image.load("/home/matteo/Downloads/Bomb Person.png").convert_alpha()
        self.rect = self.image.get_rect(center=center)
        self.speed = speed

    def update(self, target_point):
        current_point = self.rect.center
        delta = (
            target_point[0] - current_point[0],
            target_point[1] - current_point[1]
        )
        delta_len = math.sqrt(delta[0]**2 + delta[1]**2)
        if delta_len == 0:
            speed_vector = (0, 0)
        else:
            speed_vector = (
                delta[0] / delta_len * self.speed,
                delta[1] / delta_len * self.speed
            )
        self.rect.move_ip(speed_vector[0], speed_vector[1])

pygame.init()

screen = pygame.display.set_mode((500, 500))

clock = pygame.time.Clock()
FPS = 30

player = Player((
    random.randint(0, 500),
    random.randint(0, 500)
))

enemy = Enemy((
    random.randint(0, 500),
    random.randint(0, 500)
))

enemies = pygame.sprite.Group()
enemies.add(enemy)

all_sprites = pygame.sprite.Group()
all_sprites.add(player)
all_sprites.add(enemies)

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.MOUSEBUTTONDOWN:
            pointer_position = event.pos
            if enemy.rect.collidepoint(pointer_position):
                enemy.kill()

    screen.fill("white")

    pressed_keys = pygame.key.get_pressed()

    if (player.alive):
        player.update(pressed_keys)
    if (enemy.alive):
        enemy.update(player.rect.center)

    collision = pygame.sprite.spritecollide(player, enemies, False)
    if len(collision):
        player.kill()

    all_sprites.draw(screen)

    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```
