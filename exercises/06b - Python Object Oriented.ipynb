## Esercizio 1 - Car
Implementa la classe mostrata in figura:  
![Car class](https://www.plantuml.com/plantuml/svg/SoWkIImgAStDuKhEIImkLd1EB5AeviehpqXDpR5IyCmhuSfBoSylAeQmS_DJKdFYyn9pKwqKWakAClFIkVBByrABuBoC0eX2xVo2r3mDJIxjv9pyubGG0smFa57BvP2QbmAq4000)
- `nwheel` è il numero di ruote
- `ndoors` è il numero di porte
- `model_name` è il nome del modello
- `owner` è il nome del proprietario
- `hp` è la potenza della macchina espressa in cavalli motore (*horse power*)


````python
class Car:
    def __init__(self, nwheel, ndoors, model_name, owner, hp):
        self.nwheel = nwheel
        self.ndoors = ndoors
        self.model_name = model_name
        self.owner = owner
        self.hp = hp

    def open(self):
        print('La macchina è aperta')

    def close(self):
        print('La macchina è chiusa')

    def start(self):
        print('La macchina è accesa')

    def __repr__(self):
        return f'nwheel={self.nwheel} ndoors={self.ndoors} model={self.model_name} owner={self.owner} hp={self.hp}'
````

````python
punto = Car(4, 5, 'Punto', 'Simone', 62)

print(f'La macchina è una {punto.model_name} e ha {punto.hp} cavalli')

punto.open()
punto.start()

print(punto)
````

## Esercizio 2 - Point
Definisci una classe Point come in figura:  
![Point class](https://www.plantuml.com/plantuml/svg/SoWkIImgAStDuKhEIImkLWZ8pymhKQZcgh1ISCl9JopXgeGnKZA1IlEIKpKAh22gD64ojLmEgNafG3C0)

*Esempio invocazione `distance`*:
```python
p1 = Point(0, 0)
p2 = Point(2, 1)
distance = p1.distance(p2)
print(f'La distanza tra p1 e p2 e\' {distance}')
```
Viene stampato:
```
La distanza tra p1 e p2 e' 5
```

````python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, p1):
        if type(p1) != Point:
            return 0

        return math.sqrt((self.x - p1.x) ** 2 + (self.y - p1.y) ** 2)
````

````python
p1 = Point(0, 0)
p2 = Point(2, 1)
distance = p1.distance(p2)
print(f'La distanza tra p1 e p2 e\' {distance}')
````
