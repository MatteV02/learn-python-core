## Esercizio 1 - max_list
Scrivi una funzione che, data una **lista** di numeri, fornisca in output il maggiore tra tutti gli elementi della lista.

La funzione deve avere la seguente firma:
```python
def max_list(list)
```
- list è una lista di elementi (ritorna `None` se alla funzione non è passata una lista)

La funzione ritorna l'elemento massimo all'interno della lista.

````python
tests = [
    {'list': [1,2,3,4], 'expected': 4},
    {'list': ['a', 'd', 'c', 'Z'], 'expected': 'd'},
    {'list': 1234, 'expected': None}
]

for test in tests:
    max = max_list(test['list'])
    if (max != test['expected']):
        print(f'TEST max_list({test['list']}): FAIL')
    else:
        print(f'TEST max_list({test['list']}): PASSED')
````

## Esercizio 2 - search_list
Scrivi un una funzione che a partire da un elemento e una lista di elementi dica in output se l'elemento passato sia presente o meno nella lista.  

Qualora l'elemento sia presente nella lista, il programma dovrà comunicarci l'indice dell'elemento tramite il metodo `index` (vedi [documentazione](https://docs.python.org/it/3.14/library/stdtypes.html#sequence.index)).

La funzione deve avere la seguente firma:
```python
def search_list(list, el)
```
- `list` è la lista su cui cercare (ritorna -1 se a questo parametro non è passata una lista)
- `el` è l'elemento da cercare

La funzione ritorna -1 se non trova l'elemento all'interno della lista, altrimenti ritorna un indice con la posizione dell'elemento.

````python
tests = [
    {'list': [1,2,3,4], 'el': 2, 'expected': 1},
    {'list': ['a', 'd', 'c', 'Z'], 'el': 'Z', 'expected': 3},
    {'list': 1234, 'el': 1, 'expected': None}
]

for test in tests:
    max = search_list(test['list'])
    if (max != test['expected']):
        print(f'TEST search_list({test['list']}, {test['el']}): FAIL')
    else:
        print(f'TEST search_list({test['list']}, {test['el']}): PASSED')
````

## Esercizio 3 - print_histogram
Scrivi una semplice funzione che, data una lista di numeri, fornisca in output un istogramma basato su questi numeri, usando asterischi per disegnarlo.

Data ad esempio la lista [3, 7, 9, 5], la funzione dovrà produrre questa sequenza:
```
***
*******
*********
*****
```

La funzione deve avere la seguente definizione:
```python
def print_histogram(values)
```
- `values` è una lista di `int` (esci senza fare nulla se non è passata alla funzione una lista di int)

````python
tests = [
    [3, 7, 9, 5],
    [1, 2, 3],
    [],
    (1, 5),
    ["a", 2]
]

for test in tests:
    print(f'print_histogram({test}):')
    print_histogram(test)
    print("---\n")
````

## Esercizio 4 - intersection
Dati 2 set di numeri, scrivere una funzione che faccia l'intersezione tra i 2 insiemi.

La funzione deve avere la seguente definizione:
```python
def intersection(setA, setB)
```
- `setA` e `setB` sono 2 set (ritorna `None` se vengono passati alla funzione degli argomenti di tipo diverso)

La funzione ritorna un seto contenente l'intersezione tra i 2 insiemi.  
*Es.*  
```python
intersection({1,2}, {2,3}) # ritorna {2}
```

````python
tests = [
    {'setA': {10, 20, 30, 40, 50}, 'setB': {30, 40, 50, 60, 70}, 'expected': {30, 40, 50}},
    {'setA': {10, 20, 30, 40, 50}, 'setB': 20, 'expected': None},
    {'setA': {10, 20}, 'setB': {30, 40, 50, 60, 70}, 'expected': {}}
]

for test in tests:
    ret = intersection(test['setA'], test['setB'])
    if (ret != test['expected']):
        print(f'TEST intersection({test['setA']}, {test['setB']}): FAIL')
    else:
        print(f'TEST intersection({test['setA']}, {test['setB']}): PASSED')
````

## Esercizio 5 - create_person

Crea una funzione che generi un dizionario contenente i dati di una persona.

La funzione ha la seguente definizione:
```python
def create_person(name, last_name, age, email)
```
con:
- `name`, `last_name` e `email` che sono delle stringhe (ritorna `None` se vengono passati dei dati diversi)
- `age` è un numero intero compreso tra 0 e 120 (ritorna `None` se viene passato un valore errato)

La funzione ritorna un dizionario contenente i dati della persona.
*Es.*
```python
create_person('Matteo', 'Veroni', 23, 'veroni.matteo@icloud.com')
```
ritorna
```
{'name': 'Matteo', 'last_name': 'Veroni', 'age': 23, 'email': 'veroni.matteo@icloud.com'}
```

````python
tests = [
    {'name': 'Matteo', 'last_name': 'Veroni', 'age': 23, 'email': 'veroni.matteo@icloud.com', 'fail': False},
    {'name': 'Matteo', 'last_name': 'Veroni', 'age': '23', 'email': 'veroni.matteo@icloud.com', 'fail': True},
]

for test in tests:
    ret = create_person(test['name'], test['last_name'], test['age'], test['email'])
    if test['fail']:
        expected = False
    else:
        test.pop('fail')
        expected = test
    if ret == expected:
        print(f'TEST create_person({test['name']}, {test['last_name']}, {test['age']}, {test['email']}): SUCCESS')
    else:
        print(f'TEST create_person({test['name']}, {test['last_name']}, {test['age']}, {test['email']}): FAIL')

````
