## Esercizio 1 - Crea e visualizza 2 sprite
Crea 2 sprite, il primo rappresenterà il personaggio comandato dall'utente, il secondo rappresenterà il nemico.

## Esercizio 2 - Update sprite
Aggiungi agli sprite precedenti una funzione update che permetta allo sprite dell'utente il movimento da tastiera, mentre lo sprite che rappresenta il nemico dovrà muoversi ad ogni istante verso l'utente.  
Se non riesci a implementare la logica del movimento automatico del nemico, non ti preoccupare. Concentrati sul fare in modo che il personaggio dell'utente si muova in base ai tasti premuti.  
<img src="resources/07b-es2.png" alt="schermata esercizio 2" width=350>  

## Esercizio 3 - Rilevamento contatto
Aggiungi una logica che permetta di determinare il contatto tra nemico e utente. Se l'utente viene toccato dal nemico, uccidi l'utente.  
Se non riesci a rimuovere l'utente dalla visualizzazione non ti preoccupare. Mostra un contatore sullo schermo che conti quante volte utente e nemico si sono toccati.  
**SUGGERIMENTO**: Aggiungi il nemico ad un gruppo di nemici e usa la funzione vista durante la lezione per rilevare la collisione tra nemico e utente.
<img src="resources/07b-es3.png" alt="schermata esercizio 3" width=350>  

## Esercizio 4 - Mouse
Aggiungi una logica che permetta di leggere i click del mouse. Se l'utente clicca sul nemico elimina il nemico.  
**SUGGERIMENTO**: puoi rilevare se un click è stato fatto sul nemico controllando se il punto su cui l'utente clicca sta dentro al rettangolo occupato dal nemico.  
Usa questa funzione:
`enemy.rect.collidepoint(point)`  
[Documentazione](https://pyga.me/docs/ref/rect.html#pygame.Rect.collidepoint)
