## Esercizio 1 - Disegna un rettangolo e un cerchio
Replica la seguente schermata:


```python
import pygame

pygame.init()

screen = pygame.display.set_mode((500, 500))

clock = pygame.time.Clock()
FPS = 30

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill("white")

    rect = pygame.Rect((10,10), (200, 480))
    pygame.draw.rect(screen, "orange", rect)

    pygame.draw.circle(screen, "green", (375, 250), 75)

    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```

## Esercizio 3 - Disegna un rombo

```python
import pygame

pygame.init()

screen = pygame.display.set_mode((500, 500))

clock = pygame.time.Clock()
FPS = 30

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill("white")

    pygame.draw.polygon(screen, "purple", [
        (250, 80),
        (490, 250),
        (250, 420),
        (10, 250)
    ])

    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```

## Esercizio 4 - Stampa una scritta a schermo
Replica la seguente schermata:


```python
import pygame

pygame.init()

screen = pygame.display.set_mode((500, 500))

clock = pygame.time.Clock()
FPS = 30

font = pygame.font.Font("/home/matteo/Downloads/ubuntu_5.2.8/ttf/ubuntu-latin-400-normal.ttf", size=40)

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill("white")

    welcome = font.render("Welcome to PyGame!", True, "black")
    screen.blit(welcome, (65, 225))

    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```

## Esercizio 5 - Stampa una foresta
Replica la seguente schermata:


```python
import pygame
import random

pygame.init()

screen = pygame.display.set_mode((500, 500))

clock = pygame.time.Clock()
FPS = 30

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill("white")

    for i in range(25):
        for j in range(25):
            tree = pygame.image.load("/home/matteo/Downloads/Tree.png").convert_alpha()
            screen.blit(tree, (
                i * 19 + 10,
                j * 19 + 10
            ))

    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```
