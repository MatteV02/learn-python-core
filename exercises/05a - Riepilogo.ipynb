## Esercizio 1 - describe_person

Scrivi una funzione che presi 3 parametri in ingresso, nome, cognome ed età ritorni una stringa che contenga la descrizione della persona: `NOME COGNOME, age: ETA`.  
La funzione deve avere la seguente firma.  
Nel caso in cui i parametri passati non siano validi la funzione ritorna una stringa vuota.  
```python
def describe_person(first_name, last_name, age):
    ...
```
- `first_name` rappresenta il nome e deve essere una stringa
- `last_name` rappresenta il cognome deve essere una stringa
- `age` rappresenta l'età e deve essere un intero positivo minore di 120

Esempio:
```python
describe_person('Anna', 'Pepe', 22)
```
ritorna `'Anna Pepe, age: 22'`


````python
# in questa cella inserisci il codice
def describe_person(first_name, last_name, age):
    if type(first_name) != str:
        return ''
    if type(last_name) != str:
        return ''
    if type(age) != int or (age < 0 or age >= 120):
        return ''
    return f'{first_name} {last_name}, age: {age}'

print(describe_person('Anna', 'Pepe', -2))
````

````python
tests = [
    {'description': 'Test valido 1', 'first_name': 'Anna', 'last_name': 'Pepe', 'age': 22, 'expected': 'Anna Pepe, age: 22'},
    {'description': 'Test valido 2', 'first_name': 'Sergio', 'last_name': 'Mattarella', 'age': 84, 'expected': 'Sergio Mattarella, age: 84'},
    {'description': 'Test valido 3', 'first_name': 'Leone Lucia', 'last_name': 'Ferragni', 'age': 7, 'expected': 'Leone Lucia Ferragni, age: 7'},
    {'description': 'Test invalido: tipo nome sbagliato', 'first_name': 5, 'last_name': 'Ferragni', 'age': 7, 'expected': ''},
    {'description': 'Test invalido: tipo cognome sbagliato', 'first_name': 'Leone Lucia', 'last_name': 5, 'age': 7, 'expected': ''},
    {'description': 'Test invalido: tipo eta sbagliato', 'first_name': 'Leone Lucia', 'last_name': 'Ferragni', 'age': '7', 'expected': ''},
    {'description': 'Test invalido: tipo eta invalida', 'first_name': 'Leone Lucia', 'last_name': 'Ferragni', 'age': 150, 'expected': ''},
]

for test in tests:
    print(f'Test: {test['description']}')
    print(f'\tfirst_name: {test['first_name']}, last_name: {test['last_name']}, age: {test['age']}')
    if describe_person(test['first_name'], test['last_name'], test['age']) == test['expected']:
        print(f'\tTEST PASSED')
    else:
        print(f'\tTEST FAILED')
````

## Esercizio 3 - print_triangle
Scrivi una funzione che dato un lato di un triangolo stampi a schermo un triangolo.  
La funzione deve avere la seguente firma.  
Nel caso in cui i parametri passati non fa nulla.  
```python
def print_triangle(n):
    ...
```
- `n` rappresenta la lunghezza del lato, deve essere un numero intero > 0

Esempio:
```python
print_triangle(2)
```
stampa
```
+
|\
| \
+--+
```

```python
print_triangle(1)
```
stampa
```
+
|\
+-+
```

````python
def print_triangle(n):
    if type(n) != int or n <= 0:
        return
    print('+')
    for i in range(n):
        space = ' ' * i
        print(f'|{space}\\')
    trattini = '-' * n
    print(f'+{trattini}+')

print_triangle(10)
````
