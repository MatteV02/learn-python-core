## Esercizio 1 - print_comments
Nella cartella resources all'interno della cartella delle esercitazioni puoi trovare il file input.txt.  
Scrivi un codice che apra il file input.txt e stampi a schermo solo le linee che cominciano col carattere `#`.

````python
import os
path = os.path.realpath('resources/input.txt')
if os.path.isfile(path):
    with open(path, 'r') as file:
        for line in file:
            if (line[0] == '#'):
                print(line.strip())
else:
    print(f'Impossibile trovare il file {path}')
````

## Esercizio 2 - write_savefile
Scrivi una funzione che dato un punteggio lo salvi sul savefile specificato all'interno della cartella resources.  
La funzione deve avere la seguente firma:
```python
def write_savefile(filename, score)
```
con:
- `filename`: il nome del file
- `score`: il punteggio da salvare
- La funzione scrive il solo punteggio ottenuto sul file di nome specificato. Non fa niente nei seguenti casi:
    - il filename non sia una stringa che termina con `.txt`
    - `score` non è un numero intero
    - `score` negativo
    - l'operazione di scrittura fallisce

````python
def write_savefile(filename, score):
    if len(filename) < 4 or filename[-4:] != '.txt':
        return
    if type(score) != int:
        return
    if score < 0:
        return
    path = os.path.realpath(f'resources/{filename}')
    with open(path, 'w') as file:
        file.write(str(score))
````

````python
# Verifica manualmente che venga salvato il numero 10 nel file savefile.txt e verifica che venga posizionato all'interno della cartella **resources** sotto exercises
write_savefile('savefile.txt', 10)
````

## Eserizio 3 - read_score_from_savefile
Scrivi una funzione che legga il contenuto del file creato al passo precedente.  
La funzione deve avere la seguente firma:
```python
def read_score_from_savefile(filename)
```
con:
- `filename`: il nome del file da leggere
- La funzione apre il file e ritorna il punteggio all'interno contenuto (si supponga il file contenga solo il numero di punteggio come creato al passo precedente). La funzione non fa nulla nei seguenti casi:
    - il filename non è una stringa che termina con `.txt`
    - l'operazione di lettura fallisce

````python
def read_score_from_savefile(filename):
    if len(filename) < 4 and filename[-4:] != '.txt':
        return
    path = os.path.realpath(f'resources/{filename}')
    if not os.path.isfile(path):
        return
    with open(path, 'r') as file:
        file_content = file.read().strip()
        score = int(file_content)
    return score
````

````python
score = read_score_from_savefile('savefile.txt')
print(f'Il punteggio letto è {score}')
````
